```python
# TASK: Create a function get_highest_degree_nodes(G) to get the node(s) with the highest degree
#       and test the function to the above graph.
# Notes: You are only allowed to use the G.degree(node) method and basic loops!

def get_highest_degree_nodes(G):
    highest_degree = 0
    highest_degree_nodes = []

    for node in G.nodes():
        degree = G.degree(node)
        
        if degree > highest_degree:
            highest_degree = degree
            highest_degree_nodes = [node]
        elif degree == highest_degree:
            highest_degree_nodes.append(node)

    return highest_degree_nodes

get_highest_degree_nodes(G)
```

```python
G = nx.Graph()
G.add_nodes_from(range(1, 8))

cluster1_edges = [(1, 2), (1, 3), (2, 3), (2, 4)]
cluster2_edges = [(5, 6), (5, 7), (6, 7)]
bridge_edge = [(4, 5)]

G.add_edges_from(cluster1_edges)
G.add_edges_from(cluster2_edges)
G.add_edges_from(bridge_edge)

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=700)
nx.draw_networkx_edges(G, pos, edgelist=cluster1_edges, edge_color='blue', width=2)
nx.draw_networkx_edges(G, pos, edgelist=cluster2_edges, edge_color='red', width=2)
nx.draw_networkx_edges(G, pos, edgelist=bridge_edge, edge_color='green', width=2)
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black', font_weight='bold')
plt.axis('off')
plt.show()
```
